```python
import tensorflow as tf

loaded_model = tf.keras.models.load_model('./vae_decoder.h5')
```

```python
import matplotlib.pyplot as plt
import numpy as np

def generate_and_save_images():
  fig = plt.figure(figsize=(16, 16))
    
  for i in range(16):
    z = tf.random.normal([1, 8])
    predictions = loaded_model(z)
    plt.subplot(4, 4, i+1)
    plt.imshow(tf.sigmoid(predictions[0]))

  plt.savefig('vae_test_output.png')
  plt.show()
              
generate_and_save_images()
```
