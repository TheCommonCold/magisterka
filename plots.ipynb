```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
import pickle

def moving_average(x, w):
    return np.convolve(x, np.ones(w), 'valid') / w
```

```python
with open('./data/combined_dataset/gan_loss.obj', 'rb') as fp:
  losses = pickle.load(fp)

plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.plot(losses['gen'], label='loss')
plt.plot(moving_average(losses['gen'], 50), label='50 epoch moving average loss')
plt.legend()
plt.savefig("gan_generator_loss.svg", format="svg")
```

```python
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.plot(losses['disc'], label='loss')
plt.plot(moving_average(losses['disc'], 50), label='50 epoch moving average loss')
plt.yscale('log')
plt.legend()
plt.savefig("gan_discriminator_loss.svg", format="svg")
```

```python
with open('./vae_loss.obj', 'rb') as fp:
  loss = pickle.load(fp)

plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.plot(loss, label='loss')
plt.plot(moving_average(loss, 50), label='20 epoch moving average loss')
plt.legend()
plt.savefig("vae_loss.svg", format="svg")
```
